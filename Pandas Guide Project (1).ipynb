### Creating Series
Series can be created by giving a list or dictionary.
<br>
In example 1, s1(series1) was created with a list. <br>
In example 2, s2(series2) was created with a dictionary. <br>
In example 3, s3(series3) was created with a list but we assigned the index directly. <br>
When it is created with a list, the index will automatically start with 0.<br>
When it is created with a dictionary, the key value will become the index.<br>

```python
s1 = pd.Series(['one','two','three','four'])
s1
#example 1 
```

```python
s2= pd.Series({
    1:'one',
    2:'two',
    3:'three',
    4:'four'
})
s2
#example 2
```

```python
s3= pd.Series([1,2,3,4], index = ['one','two','three','four'])
s3
#example 3
```

### Pulling values from Series by its index
We can pull specific values in Series. <br>
The codes below will show different ways.

```python
#This series is created as an example for this section
s4= pd.Series([10,30,50,70], index=['I','N','S','T'])
```

```python
#From the series 4, we created, this code pulls the value of the index N which is 30.
s4['N']
```

```python
#Since Series is also a form of sequence, we can also use the original index that starts by 0 to n.
s4[1]
```

```python
#Since the values are all numbers for Series4, we can also do something like this.
s4 *2
# This will multiply the values with 2.
```

```python
#You can also make conditionals for the values in the series.
s4[s4>=50]
```

### Creating DataFrames
As explained earlier, the Dataframe is Panda's primary data structure. It is divided into axes(Rows and Columns).<br>
The code below shows how to create Dataframe in different ways.

```python
# df1(Dataframe1) is created by pd.DataFrame({dictionary})
df1 = pd.DataFrame(
        {"a" : [1 ,2, 3],
        "b" : [4, 5, 6],
        "c" : [7, 8, 9]})
df1
```

```python
#df2(Dataframe2) is created by pd.DataFrame(dictionary)
two = {
    "1":[3,6,9],
    "2":[4,8,12],
    "3":[5,15,20]}
df2= pd.DataFrame(two)
df2
```

```python
# The dataframe can also be created by using series.
s5 = pd.Series(
    [1000,2000,3000], 
    index = ['2020-05-01', '2020-05-02', '2020-05-03'], 
    name='apple')

s6 = pd.Series(
    [4000,5000,6000], 
    index = ['2020-05-01', '2020-05-02', '2020-05-03'], 
    name='bannana')

s7 = pd.Series(
    [7000,8000,9000], 
    index = ['2020-05-01', '2020-05-02', '2020-05-03'], 
    name='coconut')
df3 = pd.DataFrame([s5,s6,s7])
df3
```

### Pulling values from Dataframe 

```python
#You can pull values by selecting a column.
df3["2020-05-01"]
```

```python
#You can pull values by selecting a row by using .loc[] function.  
df3.loc['apple']
```

```python
#You can pull values of a row by its index by using .iloc[] function.
df3.iloc[1]
```

```python
#It is also possible to use conditions for both rows and columns to look for data you want to find.
df3.loc['apple':'coconut', '2020-05-02':'2020-05-03']
```

```python
#You can also use conditions.
df3.loc[df3['2020-05-03']>4000,:]
```

### Data Visualization

There are many different types of data that data science deals with.
The size of the data is often thousands or tens of thousands.<br>
Here's why you use visualization tools. If you use the graphs, it is practical to grasp the structure and pattern of the data. It also provides insight into data from a variety of perspectives.<br>
Pandas library also offers data visualization but it may not be the best library for data visualization. <br>
Matplotlib supports various formats and functions regarding 2D plan graphs to the extent that they can be called Python standard visualization tools. There are different libraries other than Matplotlib such as SeaBorn. I will use Matplotlib for example.
<br> To use Matplotlib, We are going to import the library.


```python
import matplotlib.pyplot as plt
```

We are going to create a line graph with the NBA dataset.<br>
I am going to make a line graph of Chicago Bulls 19-20 season goal points. <br>
Data visualization Example<br>
I first filtered the columns: 'season' and 'team_abbreviation'. <br>
I used the 'player_name' as index which automatically turns into x axis and 'pts' column as y axis. <br>
I used .xticks function to make the. labels for x to be vertical. <br>
I used .title .xlabel .ylabel to label the title, x label, and y label. <br>
I used .legend to make the 2019-20 Chicago Bulls Points in the graph.

```python
season1920new=data.loc[(data['season'] == '2019-20') & (data['team_abbreviation']== 'CHI')]
season1920new
chicago_1920= season1920new[['player_name','team_abbreviation','pts']]
chicago_1920.index=season1920new.player_name
plt.plot(chicago_1920.pts)
plt.xticks(rotation='vertical')
plt.title('2019-20 Chicago Bulls Points')
plt.xlabel('Names')
plt.ylabel('Points')
plt.legend(labels=['2019-20 Chicago Bulls Points'], loc='best')
```
